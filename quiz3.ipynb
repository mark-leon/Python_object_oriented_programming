```python
my_tuple = (("Firmino", 1101),("Gabriel", 1201),("Casemiro", 1302),("Fred", 
1402),("Messi", 1111),("Aguero", 1211),("Klose", 1121), ("Werner", 1221), 
("Toni", 1322))


print("The original tuple : " + str(my_tuple))
  
s = [{'key': sub[0], 'value': sub[1],} 
      for sub in my_tuple] 

print("The converted dictionary : " + str(s))
b = len(s)
a = []
c = []
d = []
for i in range(b):
  a.append(s[i]['value'])
for x in a:  
  n = x%10
  if n==1:
    c.append("forward") 
  elif n==2:
    c.append("midfield")     
for x in a:
  m = int(x/10)%10
  if m == 0:
    d.append("Brazil") 
  elif m == 1:
    d.append("Argentina") 
  elif m == 2:
    d.append("Germany")
print(c)
print(d)
```
