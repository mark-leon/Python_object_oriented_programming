```python

class Student():

  def __init__(self,name = "default"):
    self.count = 0
    self.name = name
    self.prevCourse = None
    if name =="default":
      print("A course can not be added without student name")
    
  def addCourse(self,course,department):
    self.course = course
    if self.course ==self.prevCourse:
      self.count = self.count
    else:
       self.count = self.count+1
    self.prevCourse = course
    self.department = department

  def setName(self,name):
    self.name = name

   
  def printDetail(self):
    print("\nNo of courses:",self.count,
          "\nStudent name:",self.name)
    print(self.department+":"+self.course)








s1 = Student()
print("=================================")
s1.addCourse("CSE110", "Computer Science")
print("=================================")
s1.setName("Bob")
s1.addCourse("CSE110", "Computer Science")
s1.printDetail()
print("=================================")
s2 = Student("Carol")
s2.addCourse("CSE111", "Computer Science")
s2.addCourse("MAT110", "Mathematics")
print("=================================")
s2.printDetail()
s2.addCourse("CSE230", "Computer Science")
print("=================================")
s2.printDetail()
print("=================================")







# A course can not be added without student name
# =================================
# Number of Course(s): 1
# Student Name: Bob
# Computer Science: CSE110
# =================================
# =================================
# Number of Course(s): 2
# Student Name: Carol
# Computer Science: CSE111
# Mathematics: MAT110
# =================================
# Number of Course(s): 3
# Student Name: Carol
# Computer Science: CSE111, CSE230
# Mathematics: MAT110
# =================================













```
