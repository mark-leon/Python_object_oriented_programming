```python
###1




class Student:
  def __init__(self, name='Just a student', dept='nothing'):
    self.__name = name
    self.__department = dept
  def set_department(self, dept):
    self.__department = dept
  def get_name(self):
    return self.__name
  def set_name(self,name):
    self.__name = name

  def __str__(self):
    return 'Name: '+self.__name+' Department: '+self.__department


class BBA_Student(Student):
   def __init__(self,name="Default",dept="BBA"):
     Student.__init__(self,name,dept)


print(BBA_Student())
print(BBA_Student('Humpty Dumpty'))
print(BBA_Student('Little Bo Peep'))







###2

class Vehicle:
  def __init__(self):
    self.x = 0
    self.y = 0
  def moveUp(self):
    self.y+=1
  def moveDown(self):
    self.y-=1
  def moveRight(self):
    self.x+=1
  def moveLeft(self):
    self.x-=1
  def __str__(self):
    return '('+str(self.x)+' , '+str(self.y)+')'


class Vehicle2010(Vehicle):
  def moveLowerLeft(self):
    self.x-=1
    self.y-=1
  def moveUpperLeft(self):
    self.x-=1
    self.y+=1
  def moveUpperRight(self):
    self.x+=1
    self.y+=1
  def moveLowerRight(self):
    self.x+=1
    self.y-=1
  def equals(self,a):
    if car1.x == a.x and car1.y == a.y:
      return True
    else:
      return False



print('Part 1')
print('------')
car = Vehicle()
print(car)
car.moveUp()
print(car)
car.moveLeft()
print(car)
car.moveDown()
print(car)
car.moveRight()
print(car)
print('------')
print('Part 2')
print('------')
car1 = Vehicle2010()
print(car1)
car1.moveLowerLeft()
print(car1)
car2 = Vehicle2010()
car2.moveLeft()
print(car1.equals(car2))
car2.moveDown()
print(car1.equals(car2))
print('------')






###3






class Tournament:
  def __init__(self,name):
    self.__name = name
  def set_name(self,name):
    self.__name = name
  def get_name(self):
    return self.__name



class Cricket_Tournament(Tournament):
  def __init__(self,name ='Default',no_teams = 0,type="No type"):
    Tournament.__init__(self,name)
    self.no_teams = no_teams
    self.type = type
  def detail(self):
    return "\nCricket Tournament Name:"+super().get_name()+"\nNumber of Teams: "+str(self.no_teams)+"\nType:"+self.type


class Tennis_Tournament(Tournament):
  def __init__(self,name,no_players):
    Tournament.__init__(self,name)
    self.no_players = no_players
  def detail(self):
    return "\nTennis Tournament Name: "+super().get_name()+"\nNumber of Players: "+str(self.no_players)
  

ct1 = Cricket_Tournament()
print(ct1.detail())
print("-----------------------")
ct2 = Cricket_Tournament("IPL",10,"t20")
print(ct2.detail())
print("-----------------------")
tt = Tennis_Tournament("Roland Garros",128)
print(tt.detail())






###4

class Product:
  def __init__(self,id, title, price):
    self.__id = id
    self. __title = title
    self. __price = price

  def get_id_title_price(self):
    return "ID: "+str(self.__id)+" Title: "+self.__title+" Price:"+str(self.__price)

class Book(Product):
  def __init__(self,id,title,price,Isbn,publisher):
    Product.__init__(self,id,title,price)
    self.Isbn = Isbn
    self.publisher = publisher

  def printDetail(self):
    return super().get_id_title_price()+"\nISBN: "+self.Isbn+" Publisher: "+self.publisher


class CD(Product):
  def __init__(self,id,title,price,band,duration,genre):
    Product.__init__(self,id,title,price)
    self.band = band
    self.duration = duration
    self.genre = genre

  def printDetail(self):
    return super().get_id_title_price()+"\nBand: "+self.band+" Duration: "+str(self.duration)+"\nGenre: "+self.genre


book = Book(1,"The Alchemist",500,"97806","HarperCollins")
print(book.printDetail())
print("-----------------------")
cd = CD(2,"Shotto",300,"Warfaze",50,"Hard Rock")
print(cd.printDetail())











###5


class Animal:
  def __init__(self,sound):
    self.__sound = sound
  def makeSound(self):
    return self.__sound

class Printer:
  def printSound(self, a):
    print(a.makeSound())

class Dog(Animal):
  def __init__(self,sound):
    Animal.__init__(self,sound)

class Cat(Animal):
  def __init__(self,sound):
    Animal.__init__(self,sound)

 
d1 = Dog('bark')
c1 = Cat('meow')
a1 = Animal('Animal does not make sound')
pr = Printer()
pr.printSound(a1)
pr.printSound(c1)
pr.printSound(d1)










###6



class Shape:
  def __init__(self, name='Default', height=0, base=0):
    self.area = 0
    self.name = name
    self.height = height
    self.base = base
  def get_height_base(self):
    return "Height: "+str(self.height)+",Base: "+str(self.base)


class triangle(Shape):
  def __init__(self,*value):
    if len(value) == 0:
      Shape.__init__(self)
    elif len(value) == 3:
      Shape.__init__(self,value[0],value[1],value[2])
  def calcArea(self):
    self.area = 0.5*self.base*self.height
  def printDetail(self):
    return "\nShape Name: "+self.name+"\n"+super().get_height_base()+"\nArea: "+str(self.area)

  

class trapezoid(Shape):
  def __init__(self,name,height,base,side):
    Shape.__init__(self,name,height,base)
    self.side = side
  def calcArea(self):
    self.area = ((self.base+self.side)/2)*self.height
  def printDetail(self):
    return "\nShape Name: "+self.name+"\n"+super().get_height_base()+" Side_A: "+str(self.side)+"\nArea: "+str(self.area)


tri_default = triangle()
tri_default.calcArea()
print(tri_default.printDetail())
print('--------------------------')
tri = triangle('Triangle', 10, 5)
tri.calcArea()
print(tri.printDetail())
print('---------------------------')
trap = trapezoid('Trapezoid', 10, 6, 4)
trap.calcArea()
print(trap.printDetail())








###7

class Football:
  def __init__(self, team_name, name, role):
    self.__team = team_name
    self.__name = name
    self.role = role
    self.earning_per_match = 0
  def get_name_team(self):
    return 'Name: '+self.__name+', Team Name: ' +self.__team



class Player(Football):
  def __init__(self,team_name,name,role,goal,play):
    Football.__init__(self,team_name,name,role)
    self.goal = goal
    self.play = play
    self.ratio = 0
  
  def calculate_ratio(self):
    self.ratio = self.goal/self.play
    self.earning_per_match = (self.goal * 1000) + (self.play * 10)
  
  def print_details(self):
    print(super().get_name_team(),
          "\nTeam Role:", self.role,
          "\nTotal Goal:", self.goal, "Total Played:", self.play,
          "\nGoal Ratio:", self.ratio,
          "\nMatch Earning:", str(self.earning_per_match)+"k")


class Manager(Football):
  def __init__(self,team_name,name,role,win):
    Football.__init__(self,team_name,name,role)
    self.win = win
  

  def print_details(self):
    self.earning_per_match = self.win*1000 
    print(super().get_name_team(),
          "\nTeam Role:", self.role,
          "\nTotal Win:", self.win,
          "\nMatch Earning:", str(self.earning_per_match)+"k")
   

 

player_one = Player('Juventus', 'Ronaldo', 'Striker', 25, 32)
player_one.calculate_ratio()
player_one.print_details()
print('------------------------------------------')
manager_one = Manager('Real Madrid', 'Zidane', 'Manager', 25)
manager_one.print_details()
```
