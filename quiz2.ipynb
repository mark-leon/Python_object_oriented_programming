```python

class Laptop:
  def __init__(self, name=None):
    self.name =name
    self.a = []


  def add_configuration(self,*value):
    if self.name ==None:
      print("You must set name of laptop first before adding configuration")
    if len(value) ==1 and self.name !=None:
      self.a.append(value[0])
    elif len(value) ==2 and self.name !=None:
      self.a.append(value[0])
      self.a.append(value[1])
    elif len(value) ==3 and self.name !=None:
      self.a.append(value[0])
      self.a.append(value[1])
      self.a.append(value[2])
    else:
      pass

  def printDetail(self):
    if self.name ==None:
      print("Name of laptop is not set")
    else:
      print("\nLaptop:",self.name)
    self.a = list(dict.fromkeys(self.a))
    if self.name=="Dell":
      print("Configuration:",self.a)
      print("Other brands:")
      print("HP")
      print("Lenovo")
    elif self.name=="HP":
      print("Configuration:",self.a)
      print("Other brands:")
      print("Dell")
      print("Lenovo")
    elif self.name=="Lenovo":
      print("Configuration:",self.a)
      print("Other brands:")
      print("Dell")
      print("HP")
    else:
      pass
    
    
  def setName(self,new):
    self.name = new
    if self.name != "Dell" and self.name != "HP" and self.name!= "Lenovo":
      print("""Invalid name. Name must be among "Dell", "HP" or "Lenovo""")
  


dell = Laptop()
print('----------------------------------')
dell.printDetail()
print('----------------------------------')
dell.add_configuration("16 GB RAM", "Core i9")
print('----------------------------------')
dell.setName("Asus")
print('----------------------------------')
dell.setName("Dell")
dell.add_configuration("Core i9")
dell.printDetail()
print('----------------------------------')
dell.add_configuration("16 GB RAM", "1 TB SSD")
dell.printDetail()
print("===================================")
hp = Laptop("HP")
hp.add_configuration("Core i7", "8 GB RAM", "RTX 3070")
hp.printDetail()
print("===================================")
lenovo = Laptop("Lenovo")
lenovo.add_configuration("Ryzen 7")
lenovo.printDetail()


```
