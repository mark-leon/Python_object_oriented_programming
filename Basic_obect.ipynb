```python
##1

class Calculator:
  def __init__(self):
    print("Let's Calculate!")
  def add(self,value1,value2):
    print(value1+value2)
  def substract(self,value1,value2):
    print(value1-value2)
  def multiply(self,value1,value2):
    print(value1*value2)
  def divide(self,value1,value2):
    print(value1/value2)

cal = Calculator()
v1 = float(input("Value 1:"))
op = input("Operator:")
v2 = float(input("Value 2:"))
if op=="+":
  cal.add(v1,v2)
elif op=="-":
  cal.substract(v1,v2)
elif op=="*":
  cal.multiply(v1,v2)
elif op=="/":
  cal.divide(v1,v2)
else:
  pass





##2
class Customer:
  def __init__(self,name):
    self.name = name
    
  def greet(self,name = None):
    if name == None:
      print("Hello!")
    else:
      print("Hello "+self.name+"!")

  def purchase(self,*item):
    total = 0
    for x in item:
      total = total+1
    print(self.name,",","You have purchased ",str(total),"item(s)")
    for x in item:
      print(x)


customer_1 = Customer("Sam")
customer_1.greet()
customer_1.purchase("chips", "chocolate", "orange juice")
print("-----------------------------")
customer_2 = Customer("David")
customer_2.greet("David")
customer_2.purchase("orange juice")






##3
class Panda: 
  def __init__(self,name,gender,age):
    self.name = name
    self.gender = gender
    self.age = age


  def sleep(self,hours = 0):
    if hours >=3 and hours< 6:
      return self.name+" sleeps "+str(hours)+" hours daily and should have Mixed veggies"
      
    elif hours >= 6 and hours< 9:
      return self.name+" sleeps "+str(hours)+" hours daily and should have Eggplant and Tofu"
    
    elif hours >=9 and hours<=11:
      return self.name+" sleeps "+str(hours)+" hours daily and should have Broccoli Chicken"
    elif hours == 0:
      return self.name+" duration is unknown thus should have only bamboo leaves"
    else:
      pass


panda1 = Panda("Kunfu","Male", 5)
panda2=Panda("Pan Pan","Female",3)
panda3=Panda("Ming Ming","Female",8)
print("""{} is a {} Panda Bear who is {} years
old""".format(panda1.name,panda1.gender,panda1.age))
print("""{} is a {} Panda Bear who is {} years
old""".format(panda2.name,panda2.gender,panda2.age))
print("""{} is a {} Panda Bear who is {} years
old""".format(panda3.name,panda3.gender,panda3.age))
print("===========================")
print(panda2.sleep(10))
print(panda1.sleep(4))
print(panda3.sleep())







##4
class Cat:
  def __init__(self,*info):
    if len(info) == 0:
      self.color = "White"
      self.action = None
     
    elif len(info) == 1:
      self.color = info[0]
      self.action = None

    elif len(info) == 2:
      self.color = info[0]
      self.action = info[1]
    else:
      pass

  
  def printCat(self):

    if self.action==None:
      print(self.color,"Cat is sitting")
    else:
      print(self.color,"Cat is",self.action)
  
  def changeColor(self,color):
    self.color = color



c1 = Cat()
c2 = Cat("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()






##5
class Student():
  def __init__(self,name = "default"):
    self.name = name
    
  def quizcalc(self,*nums):
    sum = 0
    for x in nums:
      sum = sum+x
    self.avg = sum/3
  def printdetail(self):
    if self.name == "default":
      print("Hello",self.name,"Student")
      print("Your avg quiz score is",self.avg)
    else:
      print("Hello",self.name)
      print("Your avg quiz score is",self.avg)


s1 = Student()
s1.quizcalc(10)
print('--------------------------------')
s1.printdetail()
s2 = Student('Harry')
s2.quizcalc(10,8)
print('--------------------------------')
s2.printdetail()
s3 = Student('Hermione')
s3.quizcalc(10,9,10)
print('--------------------------------')
s3.printdetail()






##5
class Vehicle():
  def __init__(self):
    self.x = 0
    self.y = 0
  def print_position(self):
    print('(',self.x,self.y,')')
  def moveUp(self):
    self.y = self.y+1
  def moveDown(self):
    self.y = self.y-1
  def moveRight(self):
    self.x = self.x+1
  def moveLeft(self):
    self.x = self.x-1


car = Vehicle()
car.print_position()
car.moveUp()
car.print_position()
car.moveLeft()
car.print_position()
car.moveDown()
car.print_position()
car.moveRight()





##6
class Programmer():
  def __init__(self,name,language,exp):
    self.name = name
    self.language = language
    self.exp = exp
    print("Horray! A new programmer is born")
  def printDetails(self):
    print("\nName:",self.name,
          "\nLanguage:",self.language,
          "\nExperience:",self.exp,"years")
  def addExp(self,updateExp):
    self.exp = updateExp
    print("Updating the experience of",self.name)


p1 = Programmer('Ethen Hunt','Java',10)
p1.printDetails()
print('--------------------------')
p2 = Programmer("James Bond", "C++", 7)
p2.printDetails()
print('--------------------------')
p3 = Programmer("Jon Snow", "Python", 4)
p3.printDetails()
p3.addExp(5)
p3.printDetails()






##7
class Student:
  def __init__(self,*value):
    if len(value) == 2:
      self.name = value[0]
      self.id = value[1]
      self.department = "CSE"
    elif len(value) == 3:
      self.name = value[0]
      self.id = value[1]
      self.department = value[2]
    else:
      pass
  
  def dailyEffort(self,hours):
    self.hours = hours

  def printDetails(self):

    print("\nName:",self.name,
          "\nID:",self.id,
          "\nDepartment:",self.department,
          "\nDaily Effort:",self.hours)
    if self.hours <= 2:
      print("Suggestion:","Should give more effort!")
    elif self.hours <= 4:
      print("Suggestion:","Keep up the good work!")
    else:
      print("Suggestion:","Excellent! Now motivate others.")

harry = Student('Harry Potter', 123)
harry.dailyEffort(3)
harry.printDetails()
print('========================')
john = Student("John Wick", 456, "BBA")
john.dailyEffort(2)
john.printDetails()
print('========================')
naruto = Student("Naruto Uzumaki", 777, "Ninja")
naruto.dailyEffort(6)
naruto.printDetails()




##9
class Patient:
  def __init__(self,name,age):
    self.name = name
    self.age = age
  def add_Symptom(self,*symp):
    self.symp  = symp
  def printPatientDetail(self):
    print("\nName:",self.name,
          "\nAge:",self.age,
          "\nSymptoms:",self.symp)


p1 = Patient("Thomas", 23)
p1.add_Symptom("Headache")
p2 = Patient("Carol", 20)
p2.add_Symptom("Vomiting", "Coughing")
p3 = Patient("Mike", 25)
p3.add_Symptom("Fever", "Headache", "Coughing")
print("=========================")
p1.printPatientDetail()
print("=========================")
p2.printPatientDetail()
print("=========================")
p3.printPatientDetail()
print("=========================")




##10
class Avengers:
  def __init__(self,name,partner):
    self.name = name
    self.partner = partner

  def super_powers(self,*power):
    self.power = power 

  def printAvengersDetail(self):
    print("\nName:",self.name,
          "\nPartner:",self.partner,
          "\nSuper Powers:",self.power)

a1 = Avengers('Captain America', 'Bucky Barnes')
a1.super_powers('Stamina', 'Slowed ageing')
a2 = Avengers('Doctor Strange', 'Ancient One')
a2.super_powers('Mastery of magic')
a3 = Avengers('Iron Man', 'War Machine')
a3.super_powers('Genius level intellect', 'Scientist ')
print("=========================")
a1.printAvengersDetail()
print("=========================")
a2.printAvengersDetail()
print("=========================")
a3.printAvengersDetail()
print("=========================")




##11
class Shinobi:
  def __init__(self,name,rank):
    self.name = name
    self.rank = rank
    self.mission = 0
    self.salary = 0

  def calSalary(self,mission):
    self.mission = mission
    if self.rank == "Geinin":
      self.salary = self.mission*50
    elif self.rank == "Chunin":
      self.salary = self.mission*100
    else:
      self.salary = self.mission*500


  def printInfo(self):
    print("\nName:",self.name,
          "\nRank:",self.rank,
          "\nNo of mission:",self.mission,
          "\nSalary:",self.salary)


  def changeRank(self,rank):
    self.rank = rank



naruto = Shinobi("Naruto", "Genin")
naruto.calSalary(5)
naruto.printInfo()
print('====================')
shikamaru = Shinobi('Shikamaru', "Genin")
shikamaru.printInfo()
shikamaru.changeRank("Chunin")
shikamaru.calSalary(10)
shikamaru.printInfo()
print('====================')
neiji = Shinobi("Neiji", "Jonin")
neiji.calSalary(5)
neiji.printInfo()




##12
class ParcelKoro():

  def __init__(self,*value):
    if len(value) == 0:
      self.cus_name = "No name set"
      self.product_weight = 0
      self.total_fee = 0
    elif len(value) == 1:
      self.cus_name = value[0]
      self.product_weight = 0
      self.total_fee = 0
    elif len(value) == 2:
      self.cus_name = value[0]
      self.product_weight = value[1]
      self.total_fee = 0
    else:
      pass

  def calculateFee(self,*location):
    if self.product_weight == 0:
      self.total_fee = 0
    elif len(location) == 0:
      self.total_fee = (self.product_weight * 20) + 50
    elif len(location) == 1:
      self.total_fee = (self.product_weight * 20) + 100
    else:
      pass
    

  def printDetails(self):
    print("\nCustomer Name:",self.cus_name,
          "\nProduct weight:",self.product_weight,
          "\nTotal Fee:",self.total_fee)




print("**********************")
p1 = ParcelKoro()
p1.calculateFee()
p1.printDetails()
print("**********************")
p2 = ParcelKoro('Bob The Builder')
p2.calculateFee()
p2.printDetails()
print("----------------------------")
p2.product_weight = 15
p2.calculateFee()
p2.printDetails()
print("**********************")
p3 = ParcelKoro('Dora The Explorer', 10)
p3.calculateFee('Dhanmondi')
p3.printDetails()





##13
class Batsman():
  def __init__(self,*value):
    if len(value)==2:
      self.name = "New Batsman"
      self.run_scored = value[0]
      self.ball_faced = value[1]
    elif len(value)==3:
      self.name = value[0]
      self.run_scored = value[1]
      self.ball_faced = value[2]
    else:
      pass
    
  def printCareerStatistics(self):
    print("\nName:",self.name,
          "\nRun Scored:",self.run_scored,",Ball faced:",self.ball_faced)
  
  def battingStrikeRate(self):
    self.strike_rate = (self.run_scored/self.ball_faced) *100
    return self.strike_rate
  
  def setName(self,new_name):
    self.name = new_name
    

b1 = Batsman(6101, 7380)
b1.printCareerStatistics()
print("============================")
b2 = Batsman("Liton Das", 678, 773)
b2.printCareerStatistics()
print("----------------------------")
print(b2.battingStrikeRate())
print("============================")
b1.setName("Shakib Al Hasan")
b1.printCareerStatistics()
print("----------------------------")
print(b1.battingStrikeRate())





##14
class EPL_Team:
  def __init__(self,*value):
    if len(value)==1:
      self.name = value[0]
      self.slogan = "No Slogan"
      self.title = 0
    elif len(value)==2:
      self.name = value[0]
      self.slogan = value[1]
      self.title = 0
    else:
      pass
    
  def showClubInfo(self):
    return "\nName:"+self.name +"\nSong:"+self.slogan+"\nTotal No of title:"+str(self.title)
    
  def changeSong(self,cha_song):
    self.slogan = cha_song

  def increaseTitle(self):
    self.title = self.title+1



manu = EPL_Team('Manchester United', 'Glory Glory Man United')
chelsea = EPL_Team('Chelsea')
print('===================')
print(manu.showClubInfo())
print('##################')
manu.increaseTitle()
print(manu.showClubInfo())
print('===================')
print(chelsea.showClubInfo())
chelsea.changeSong('Keep the blue flag flying high')
print(chelsea.showClubInfo())






##15
class Account():
  def __init__(self,*value):
    if len(value) ==0:
      self.name = "Default Account"
      self.balance = 0.0
    elif len(value) ==1:
      self.name = value[0]
      self.balance = 0.0
    elif len(value) ==2:
      self.name = value[0]
      self.balance = value[1]
    else:
      pass
  
  def details(self):
    return "\n"+self.name+"\n"+str(self.balance)


  def withdraw(self,withdrawAmount):
    newAmount = self.balance - withdrawAmount
    if newAmount > 3070:
      print("Withdraw successful! New balance is:"+str(newAmount))
    else:
      print("Sorry, Withdraw unsuccessful! The account balance after deducting withdraw amount is equal to or less than minimum.")




a1  = Account()
print(a1.details())
print("------------------------")
a1.name = "Oliver"
a1.balance = 10000.0
print(a1.details())
print("------------------------")
a2 = Account("Liam")
print(a2.details())
print("------------------------")
a3 = Account("Noah",400)
print(a3.details())
print("------------------------")
a1.withdraw(6930);
print("------------------------")
a2.withdraw(600);
print("------------------------")
a1.withdraw(6929)






##16
class Author():
  def __init__(self,*value):
  
    if len(value) == 0:
      self.name = "Default"
    elif len(value) == 1:
      self.name = value[0]
    elif len(value) == 2:
      self.name = value[0]
      self.addBooks(value[1])
    elif len(value) == 3:
      self.name = value[0]
      self.addBooks(value[1],value[2])
    else: 
      pass
  
  def addBooks(self,*values):
    self.a = len(values)
    if len(values) ==1:
      self.book1 = values[0]
    elif len(values) ==2:
      
      self.book1 = values[0]
      self.book2 = values[1]
    elif len(values) ==3:
      self.book1 = values[0]
      self.book2 = values[1]
      self.book3 = values[2]
    else: 
      pass
  
  def printDetails(self):
    print("\nAuthor Name:",self.name,
          '\n---------------',
          "\nList of Books:")
    if self.a ==2:
      print(self.book1)
      print(self.book2)
    elif self.a ==3:
      print(self.book1)
      print(self.book2)
      print(self.book3)
    
    
  def changeName(self,new):
    self.name = new


auth1 = Author('Humayun Ahmed')
auth1.addBooks('Deyal', 'Megher Opor Bari')
auth1.printDetails()
print('===================')
auth2 = Author()
print(auth2.name)
auth2.changeName('Mario Puzo')
auth2.addBooks('The Godfather', 'Omerta', 'The Sicilian')
print('===================')
auth2.printDetails()
print('===================')
auth3 = Author('Paolo Coelho', 'The Alchemist', 'The Fifth Mountain')
auth3.printDetails()

```
